```python
sns.kdeplot(df.salary[ df.gender=="M"])
sns.kdeplot(df.salary[ df.gender=="F"])
plt.legend(["Male", "Female"])
plt.xlabel("Salary(100k)")
plt.show()
#SHOWS SALARY OF STUDENTS BASED ON GENDER
```

```python
sns.countplot("gender", hue="status", data=df)
plt.show()
# SHOWS PLACEMENT OF STUDENTS WRT GENDER
```

```python
plt.figure(figsize =(18,6))
sns.violinplot("salary", "gender", data=df)
plt.show()
```

```python
# - CONCLUSION
# SALARY OF MALES> SALARY OF FEMALES on an average.
#40- M, 30- F are unplaced.
#MALES ARE BEING OFFERED HIGHER PACKAGES OF SALARY
```

```python
sns.kdeplot(df.hsc_p[ df.status=="Placed"])
sns.kdeplot(df.hsc_p[ df.status=="Not Placed"])
plt.legend(["Placed", "Not Placed"])
plt.xlabel("Higher Secondary Education Percentage")
plt.show()
#SHOWS RELATION B/W CLASS 12TH RESULTS AND PLACEMENTS
```

```python
plt.figure(figsize =(18,8))
sns.scatterplot("salary", "hsc_b", data=df, color='m')
plt.show()
```

```python
plt.figure(figsize =(18,6))
sns.boxplot("salary", "hsc_s", data=df)
plt.show()
#SHOWS THE AFFECT OF DIFFERENT EDU.BACKGROUNDS ON SALARY
```

```python
sns.countplot("hsc_s", hue="status", data=df)
plt.show()
```

```python
sns.lineplot("hsc_p", "salary", hue="hsc_b", data=df)
plt.show()
```

```python
# - CONCLUSION
#Students of Commerce are better placed.
# Board of Education doesn't affect placement status much.
# Students with more than 70% are placed better and have higher packages.
# No conclusion of Arts Students as samples are few, they have lower salaries on an average.
```
